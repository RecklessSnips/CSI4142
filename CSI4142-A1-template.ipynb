#### Insight 10 Average price trend from Jan 2016 to Sept 2017

```python
# Make the day first for formatting (3/12/2016) and transfer to datetime.
d2['Date'] = pd.to_datetime(d2['Date'], dayfirst=True)

# First, transfer the date into month scale
# Second, group them in different month
# Third, calculate their average price
# Transfer to a new data frame: monthly_price
monthly_price = d2.groupby(d2['Date'].dt.to_period('M'))['Price'].mean().reset_index()

# get the date from monthly_price and formet them
monthly_price['Date'] = monthly_price['Date'].dt.strftime('%b %Y')

# Plot the line chart
plt.figure(figsize=(12, 6))
plt.plot(monthly_price['Date'], monthly_price['Price'], marker='o', color='b')
plt.title("Average House Price Over Time", fontsize=16)
plt.xlabel("Date (Month-Year)", fontsize=12)
plt.ylabel("Average Price (in Dollars)", fontsize=12)

# Display the price on every point (help from ChatGPT)
for i, row in monthly_price.iterrows():
    plt.text(
        i,  # x coordinate
        row['Price'],  # y coordinate
        f"{int(row['Price']):,}",  # format
        ha='center',  # align horizontally
        va='bottom',  # align vertically
        fontsize=9,  
        color='black'  
    )

# Y ticks
min_price = monthly_price['Price'].min()
max_price = monthly_price['Price'].max()
step = 50000
# Set the range
yticks = np.arange(min_price, max_price + step, step)
ytick_labels = [f"{int(y/1000)}k" for y in yticks]

plt.yticks(yticks, ytick_labels)  

plt.xticks(rotation=45, fontsize=10)
plt.grid()
plt.tight_layout()
plt.show()
```

**Insight method:** r7: Date is split into moth to expore the price. distribution

**Insight:** The housing price indicates a steady increase from February to June 2016, followed by a sharp decline in July. It then rebounded rapidly, reaching 1 million in December 2016. However, the market experienced another decline over the next seven months, lasting until July 2017. Despite this downturn, the price surged again, peaking at 1.2 million in August 2017.
